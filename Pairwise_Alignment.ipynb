```
!ls
file = input("Enter file name: ")
file = open(file)

read_file = file.read()
list_file = read_file.split()

seq1 = list_file[0]
seq2 = list_file[1]
mismatch_score = int(input("Enter the mismatch score: "))
gap_penalty = int(input("Enter the gap penalty: "))
match_score =  int(input("Enter the match score: "))
extend_penalty = gap_penalty 

seq1_list = list(seq1)
seq2_list = list(seq2)

#alignments = pairwise2.align.globalms(seq1, seq2, match_score, mismatch_score, gap_penalty, extend_penalty)

#for a in alignments: 
  #print(format_alignment(*a))
  
def match(alpha, beta):
    if alpha == beta:
        return match_score
    else:
        return mismatch_score
      
num_rows = len(seq2) + 1
num_cols = len(seq1) + 1
matrix = np.zeros(shape=(num_rows, num_cols), dtype=np.int)
str_matrix = np.zeros(shape=(num_rows, num_cols), dtype=str)

matrix[0][0] = 0
for i in range(1, num_rows):
    matrix[i][0] = matrix[i-1][0] + gap_penalty
    str_matrix[i][0] = '↑'

for j in range(1, num_cols):
    matrix[0][j] = matrix[0][j-1] + gap_penalty
    str_matrix[0][j] = '←'

seq1_list = len(list(seq1))
seq2_list = len(list(seq2))
align1 = []
align2 = []
align3 = []

str_matrix[0][0] = 0
list1 = list(range(1,len(seq2)+1))
while list1 != []:
  for i in range(1,len(seq1)+1):
        j = list1[0]
        diagonal = matrix[j - 1][i - 1] + match(seq1[i-1],seq2[j-1])
        up = matrix[j][i-1] + gap_penalty
        left = matrix[j-1][i] + gap_penalty
        max_pointer =max(diagonal, up, left)
        matrix[j][i] =  max_pointer 
        if max_pointer == diagonal:
            str_matrix[j][i] = '↖'
        elif max_pointer == up:
            str_matrix[j][i] = '↑'
        else: 
            str_matrix[j][i] = '←'
  del list1[0]

print('\n')
print('\n')
print(matrix)
print('\n')
print(str_matrix)

score = matrix[-1][-1]

print("Score: ", score)
print('\n')
                                                                                                         
i,j = seq2_list, seq1_list                                                     
while i > 0 or j > 0: 
        score_current = matrix[i][j]
        score_diagonal = matrix[i-1][j-1]
        score_up = matrix[i-1][j]
        score_left = matrix[i][j-1]
        numlist = []
        symb = []
      
        if gap_penalty == score_current - score_up:
          numlist.append(score_up)
          
        if gap_penalty == score_current - score_left:
          numlist.append(score_left)
          
        if match_score == score_current - score_diagonal:
          numlist.append(score_diagonal)
          
        if mismatch_score == score_current - score_diagonal:
          numlist.append(score_diagonal)
          
        maxnum = max(numlist)      
        if maxnum == score_diagonal:
          align1.append(seq1[j-1])
          align2.append(seq2[i-1])
          align3.append(seq2[i-1])
          i -= 1
          j -= 1  
          
        elif maxnum == score_up:
          align1.append('-')
          align2.append(seq2[i-1])
          align3.append(seq2[i-1])
          i -= 1
         
        else: 
          align1.append(seq1[j-1])
          align2.append('-')
          align3.append('-')
          j -= 1   
          
while j > 0:
        align1.append(seq1[j-1])
        align2.append('-')
        align3.append('-')
        j -= 1
while i > 0:
        align1.append('-')
        align2.append(seq2[i-1])
        align3.append(seq2[i-1])
        i -= 1
      
(align1.reverse())
(align2.reverse())
(align3.reverse())

symb = []

for a in align1:
  b = align3[0]
  if a==b:
    symb.append("|")
  elif a == "-":
    symb.append(" ")
  elif b == "-":
    symb.append(" ")
  else:
    symb.append(".")
  del align3[0]
          
a = ("".join(symb)) 
print(*align1, sep='')
print(a)
print(*align2, sep='')
                
with open("drive/My Drive/SequenceAlignment.txt", "w") as new_file:
    new_file.write("Match Score: %s \n " % match_score)
    new_file.write('\n')
    new_file.write("Mismatch Score: %s \n " % mismatch_score)
    new_file.write('\n')
    new_file.write("Gap Penalty: %s " % gap_penalty)
    new_file.write('\n')
    new_file.write('\n')
    new_file.write("Score: %s " % score)
    new_file.write('\n')
    new_file.write('\n')
    new_file.write('\n')
    string1 = ''
    string2 = ''
    for x in align1:
        string1 += x 
    new_file.write(string1)
    new_file.write('\n')
    new_file.write(a)
    new_file.write('\n')
    for x in align2:
        string2 += x 
    new_file.write(string2)
    new_file.write('\n')
    new_file.write('\n') 
    np.savetxt(new_file, matrix, fmt='%-4d', delimiter = '|', newline='\n')
    new_file.write('\n')
    np.savetxt(new_file, str_matrix, fmt='%s', delimiter = '|', newline = '\n')    
new_file.close()
```
